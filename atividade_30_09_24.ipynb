```python
import mysql.connector as mc
import pandas as pd

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ShowDoSQL").getOrCreate()

print(spark)

print(spark.version)

from pyspark.sql.functions import *
```

```python
# Definir os parametros de conexao do mysql
host = 'xx.xxx.xxx.xxx' # IP PUBLICO DA SUA INSTANCIA DO MYSQL
user = 'root' # USUARIO
password = 'xxx' # Senha
database = 'xxx'
```

```python
# Função que ao ser chamada irá realizar a nossa consulta SQL, toda vez que
# For chamada passaremos a 'query' no parametro
def sqlQuery(consulta):
  connection = mc.connect(
    host=host,
    user=user,
    password=password,
    database=database
  ) # criar a conexão

  cursor = connection.cursor() # criar o cursor que vai realizar as acoes

  cursor.execute(consulta) # realizar a consulta

  result = cursor.fetchall() # guardar o resultado

  df = pd.DataFrame(result, columns=cursor.column_names) # criar o DataFrame

  cursor.close() # Encerrar o cursor

  connection.close() # Encerrar a conexão

  return df # retorna para fora da função o Dataframe resultante
```

```python
# 2 - Liste todos os clientes que moram na cidade de "São Paulo".
clientes_cidade_sao_paulo = sqlQuery(
    '''
      SELECT * FROM cliente
      WHERE cidade = "São Paulo"
    '''
)

display(clientes_cidade_sao_paulo)
```

```python
# 3 - Liste todos os clientes ordenados pelo nome em ordem alfabética.

clientes_ordem_alfabetica_nome = sqlQuery(
    '''
    SELECT *
    FROM cliente
    ORDER BY nome ASC
    '''
)

display(clientes_ordem_alfabetica_nome)
```

```python
# 4 - Liste todos os clientes cujo nome começa com a letra 'A'.
clientes_nome_comeca_A = sqlQuery(
    '''
    SELECT *
    FROM cliente
    WHERE nome LIKE 'A%'
    '''
)

display(clientes_nome_comeca_A)
```

```python
# 5 - Selecione os 3 primeiros registros da tabela cliente.
tres_primeiros_clientes = sqlQuery(
    '''
    SELECT *
    FROM cliente LIMIT 3
    '''
)

display(tres_primeiros_clientes)
```

```python
# 6 - Liste todos os carros que são da marca 'Ford' ou 'Fiat'.
carros_ford_ou_fiat = sqlQuery(
    '''
    SELECT *
    FROM carro
    WHERE marca = 'Ford' OR marca = 'Fiat'
    '''
)

display(carros_ford_ou_fiat)
```

```python
# 7 - Encontre todos os carros cujo valor está entre 100 e 150.
carros_ford_ou_fiat = sqlQuery(
    '''
    SELECT *
    FROM carro
    WHERE valor >= 100 AND valor <= 150
    '''
)

display(carros_ford_ou_fiat)
```

```python
# 8 - Conte quantos aluguéis foram realizados.
qtd_alugueis = sqlQuery(
    '''
    SELECT COUNT(aluguel_id) AS quantidade_aluguel
    FROM aluguel
    '''
)

display(qtd_alugueis)
```

```python
# 9 - Calcule o valor total de todos os carros disponíveis para aluguel
valor_soma_carros_disp_para_aluguel = sqlQuery(
    '''
    SELECT SUM(valor) AS soma_valor_carro_disponivel
    FROM carro
    '''
)

display(valor_soma_carros_disp_para_aluguel)
```

```python
# 10 - Calcule o valor médio dos carros.
media_valor_carros = sqlQuery(
    '''
    SELECT ROUND(AVG(valor), 2) AS media_valor_carros
    FROM carro
    '''
)

display(media_valor_carros)
```

```python
# 11 - Encontre o menor e o maior valor de aluguel de carros
max_e_min_valor_aluguel = sqlQuery(
    '''
    SELECT MAX(valor) AS maior_aluguel, MIN(valor) AS menor_aluguel
    FROM carro
    '''
)

display(max_e_min_valor_aluguel)
```

```python
# 12 - Liste os estados diferentes presentes na tabela cliente.
estados_dos_clientes = sqlQuery('SELECT DISTINCT estado FROM cliente')
display(estados_dos_clientes)
```

```python
# 13 - Agrupe os clientes por estado e conte quantos clientes existem em cada estado.
qtd_clientes_por_estado = sqlQuery(
    '''
    SELECT estado, COUNT(cliente_id) AS qtd_clientes
    FROM cliente
    GROUP BY estado
    '''
)

display(qtd_clientes_por_estado)
```

```python
# 14 - Liste os modelos de carros e o número de vezes que cada um foi alugado
modelos_carros_e_qtd_alugueis = sqlQuery(
    '''
    SELECT carro.modelo, COUNT(aluguel.aluguel_id) AS qtd_alugueis
    FROM carro
    INNER JOIN aluguel ON carro.carro_id = aluguel.carro_id
    GROUP BY carro.modelo
    '''
)

display(modelos_carros_e_qtd_alugueis)
```

```python
# 15 - Liste o nome dos clientes e o número de aluguéis que cada um fez.
clientes_qtd_alugueis = sqlQuery(
    '''
    SELECT cliente.nome, COUNT(aluguel.aluguel_id) AS qtd_alugueis
    FROM cliente
    INNER JOIN aluguel ON cliente.cliente_id = aluguel.cliente_id
    GROUP BY cliente.nome
    '''
)

display(clientes_qtd_alugueis)
```

```python
# 16 = Liste todos os aluguéis, mostrando o nome do cliente e o modelo do carro alugado.
aluguel_nome_cliente_modelo_carro = sqlQuery(
    '''
    SELECT aluguel.aluguel_id, cliente.nome AS nome_cliente, carro.modelo AS modelo_carro
    FROM aluguel
    INNER JOIN cliente ON aluguel.cliente_id = cliente.cliente_id
    INNER JOIN carro ON aluguel.carro_id = carro.carro_id
    '''
)

display(aluguel_nome_cliente_modelo_carro)
```

```python
# 17 - Encontre os clientes que alugaram o carro com modelo 'Onix'.
clientes_aluguel_modelo_onix = sqlQuery(
    '''
    SELECT cliente.nome AS clientes_que_alugaram_onix
    FROM cliente
    INNER JOIN aluguel ON cliente.cliente_id = aluguel.cliente_id
    INNER JOIN carro ON aluguel.carro_id = carro.carro_id
    WHERE carro.modelo = 'Onix'
    '''
)

display(clientes_aluguel_modelo_onix)
```

```python
# 18 - Liste os nomes dos clientes e as datas em que alugaram carros.
cliente_data_aluguel = sqlQuery(
    '''
    SELECT cliente.nome, aluguel.data_aluguel
    FROM cliente
    INNER JOIN aluguel ON cliente.cliente_id = aluguel.cliente_id
    '''
)

display(cliente_data_aluguel)
```

```python
# 19 - Encontre o nome dos clientes e os modelos de carros que alugaram.
nome_cliente_modelo_carro = sqlQuery(
    '''
    SELECT cliente.nome AS nome_cliente, carro.modelo AS modelo_carro
    FROM aluguel
    INNER JOIN cliente ON aluguel.cliente_id = cliente.cliente_id
    INNER JOIN carro ON aluguel.carro_id = carro.carro_id
    '''
)

display(nome_cliente_modelo_carro)
```

```python
# 20 - Crie uma consulta que retorne o nome dos clientes e uma coluna adicional chamada status_civil que classifica os clientes como "Solteiro" ou "Casado".
nome_status_civil_clientes = sqlQuery(
    '''
    SELECT nome,
    CASE
        WHEN estado_civil = 'Solteiro' THEN 'Solteiro'
        WHEN estado_civil = 'Casado' THEN 'Casado'
    END AS status_civil
    FROM cliente
    ORDER BY status_civil
    '''
)

display(nome_status_civil_clientes)
```
